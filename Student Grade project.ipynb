# Training our model

 Separating features and label as x andy

```python
x=df.drop(columns=['CGPA'],axis=1)
y=df['CGPA']
```

# Data Standardisation

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

x_scaled=scaler.fit_transform(x)
```

# Check for multicolliernity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['vif1']=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif['features']=x.columns
vif
```

Variance Inflation Factor is less than 5 for all columns so there is no multicollinearity

# Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,accuracy_score
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
lr=LinearRegression()
```

## Finding the best random state

```python
max_acc=0
max_rs=0

for i in range(0,200):
    x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,random_state=i,test_size=0.20)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred_lr=lr.predict(x_test)
    accuracy=r2_score(y_test,pred_lr)
    print('Testing accuracy', accuracy,'random state',i)
    
    if accuracy>max_acc:
        max_acc=accuracy
        max_rs=i
        print('max_accuracy',max_acc,'max_random_state',i)
        
```

 We found Maximum testing accuracy of 97.70 at random state 42

## Train Test Split

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_train_pred=lr.predict(x_train)
print(r2_score(y_train,y_train_pred))
```

Training Accuracy 

```python
pred_test_y=lr.predict(x_test)

print(r2_score(y_test,pred_test_y))
```

Testing Accuracy

# Finding best value for cross validation of the model

```python
Train_accuracy=r2_score(y_train,y_train_pred)
Test_accuracy=r2_score(y_test,pred_test_y)


from sklearn.model_selection import cross_val_score

for j in range(2,10):
    cv_score=cross_val_score(lr,x_scaled,y,cv=j)
    cv_mean=cv_score.mean()
    print(f"At cross fold {j} the cv score is {cv_mean} and the accuracy score in training is  {Train_accuracy} and the accuracy score in testing is {Test_accuracy}")
```

At cv=9 we have highest cv score=91.61 with training accuracy=93.05 and testing accuracy of 97.70

## Cross Validation

```python
lssscore_selected=cross_val_score(lr,x_scaled,y,cv=9).mean()
print("the cv score is ",lssscore_selected)
```

# Plotting Actual vs Predicted(Linear Regression)

```python
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=pred_test_y,color='r')
plt.plot(y_test,y_test,color='b')
plt.xlabel('Actual CGPA',fontsize=14)
plt.ylabel('predicted CGPA',fontsize=14)
plt.title('Linear Regression',fontsize=18)
plt.savefig('lr.png')
plt.show()
```
